```python
import pandas as pd
import numpy as np
```

```python
# Load CSV
file_path = "output/step-1-weeks-added.csv"
df = pd.read_csv(file_path)
df.columns = df.columns.str.strip()
df.tail()
```

```python
df[['Date']].head()
```

```python
# Ensure 'Date' column is in datetime format
# df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%Y', errors='coerce')

# Extract ISO week and year
# df['Week'] = df['Date'].dt.isocalendar().week
# df['Year'] = df['Date'].dt.year  # Add Year column

# List to store processed data
final_list = []
prev_week = None
# Process each week
for (year, week), group in df.groupby(['Year', 'Week']):
    # Add the week's data
    final_list.append(group)

    # Calculate Weekly High & Weekly Low
    weekly_high = group['HIGH'].max()
    weekly_low = group['LOW'].min()

    # Create a summary row
    summary_row = pd.DataFrame({
        'Date': [""], 
        'Day': ["WEEKLY SUMMARY"], 
        'HIGH': [weekly_high], 
        'LOW': [weekly_low],
        'Year': [year],
        'Week': [week]
    })

    # Append the summary row
    final_list.append(summary_row)

# Combine all data
final_df = pd.concat(final_list, ignore_index=True)
df = final_df  # Update df with final processed data
# Sort by Year first, then by Week
df = df.sort_values(by=['Year', 'Week'], ascending=[True, True])

# Reset index after sorting
df = df.reset_index(drop=True)

df.head(20)  # Display the last 20 rows
```

```python
# Define new columns with NaN values
new_columns = ["Order Type", "Buy Quantity", "Investment", "Total Share", "Total Investment", 
               "Average Price", "Target", "IRLower", "IRUpper", "Description"]
# Columns to drop
columns_to_drop = ["series", "ltp", "vwap", "52W H", "52W L", "VOLUME", "VALUE", "No of trades"]

# Drop the columns (ignoring errors if a column is missing)
df = df.drop(columns=columns_to_drop, errors='ignore')

# Add new columns with NaN
for col in new_columns:
    df[col] = pd.NA

df.head(20)
# df.columns
```

```python
# Save the processed CSV
output_path = "output/step-2-weekly-highs-and-lows.csv"
df.to_csv(output_path, index=False)

print(f"CSV processed and saved as {output_path}.")
```
