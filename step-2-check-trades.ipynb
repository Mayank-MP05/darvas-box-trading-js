```python
import pandas as pd
import numpy as np
```

```python
# Load CSV
file_path = "output/step-1-weeks-added.csv"
df = pd.read_csv(file_path)
df.columns = df.columns.str.strip()
df.tail()
```

```python
# List to store processed data
final_list = []
prev_week = None

# Process each week
for week, group in df.groupby('Week'):
    # Add the week's data
    final_list.append(group)

    # Calculate Weekly High & Weekly Low
    weekly_high = group['HIGH'].max()
    weekly_low = group['LOW'].min()

    # Create a summary row
    summary_row = pd.DataFrame({
        'Date': [""], 
        'Day': ["WEEKLY SUMMARY"], 
        'HIGH': [weekly_high], 
        'LOW': [weekly_low]
    })

    # Append the summary row
    final_list.append(summary_row)

# Combine all data
final_df = pd.concat(final_list, ignore_index=True)
final_df.tail(20)
df = final_df
```

```python
# Define new columns with NaN values
new_columns = ["Order Type", "Buy Quantity", "Investment", "Total Share", "Total Investment", 
               "Average Price", "Target", "IRLower", "IRUpper", "Description"]
# Columns to drop
columns_to_drop = ["series", "ltp", "vwap", "52W H", "52W L", "VOLUME", "VALUE", "No of trades"]

# Drop the columns (ignoring errors if a column is missing)
df = df.drop(columns=columns_to_drop, errors='ignore')

# Add new columns with NaN
for col in new_columns:
    df[col] = pd.NA

df.head(20)
# df.columns
```

```python
# Save the processed CSV
output_path = "output/step-2-weekly-highs-and-lows.csv"
df.to_csv(output_path, index=False)

print(f"CSV processed and saved as {output_path}.")
```
